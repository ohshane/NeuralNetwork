```python
import torch
import torch.nn as nn
import torch.functional as F
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import *
```

```python
X_source, y = make_blobs(n_samples=4000,
                  n_features=2,
                  centers=4,
                  cluster_std=1.0,
                  center_box=(-5.0, 5.0),
                  shuffle=True,
                  random_state=42,
                  return_centers=None)

X_target = np.matmul(X_source, np.array([
    [0.8, 0.5],
    [0.0, 1.1]
]))

plt.figure(figsize=(8,8))
plt.scatter(X_source[:,0], X_source[:,1], c=y, marker='.', cmap='tab10', alpha=.1)
plt.scatter(X_target[:,0], X_target[:,1], c=y, marker='s', cmap='tab10', alpha=.1)
plt.axis('square')
plt.show()
```

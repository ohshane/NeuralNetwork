```python
import torch
import torch.nn as nn
import matplotlib.pyplot as plt


P = torch.FloatTensor([[1,1], [3,2], [9,1], [3,7], [7,2], [9,7], [4,8], [8,3], [1,4]])

plt.figure(figsize=(6,6))
plt.scatter(*P.T, s=100)
for i, p in enumerate(P, start=1):
    plt.annotate(rf"$p_{i}$", xy=p, size=14)
plt.axis('equal')
plt.grid()
plt.show()
```

```python
plt.pcolormesh(torch.cdist(P, P))
```

```python
import numpy as np
from scipy.cluster.hierarchy import dendrogram, linkage
from sklearn import datasets
# from fastcluster import linkage
import matplotlib.pyplot as plt

iris = datasets.load_iris()

P = torch.tensor(iris.data)
dist_mat = torch.cdist(P, P)
plt.figure(figsize=(6,6))
plt.title('distance matrix: sorted')
plt.pcolormesh(dist_mat, cmap='RdGy')
plt.show()

P = torch.tensor(iris.data[np.random.permutation(iris.data.shape[0]), :])
dist_mat = torch.cdist(P, P)
plt.figure(figsize=(6,6))
plt.title('distance matrix: arbitrarily ordered')
plt.pcolormesh(dist_mat, cmap='RdGy')
plt.show()
```

## Eigen decomposition (Symmetric)


$$
\begin{align}
A &= A^\intercal \\
A &= V \Lambda V^{-1} \\
A^\intercal &= (V \Lambda V^{-1})^\intercal \\
A^\intercal &= (V^{-1})^\intercal \Lambda^\intercal V^\intercal \\
A^\intercal &= (V^\intercal)^{-1} \Lambda V^\intercal = V \Lambda V^{-1} \quad \cdots \text{(1)(2)} \\
A &= V \Lambda V^\intercal \qquad \because V^\intercal = V^{-1} \quad \cdots \text{(5) orthogonal} \\
A &= Q \Lambda Q^\intercal
\end{align}
$$

(A^\intercal)^{-1} A^\intercal &= I_n \quad (\text{think} A^\intercal \text{as an arbitrary matrix})\\

* showing orthogonality of $\mathbf v$ in a symmetric matrix $A$

$$
\begin{align}
A &= A^\intercal \\
A \mathbf v &= \lambda_{\mathbf v} \mathbf v \\
A \mathbf w &= \lambda_{\mathbf w} \mathbf w \\

\mathbf w^\intercal A \mathbf v &= \lambda_{\mathbf v} \mathbf w^\intercal \mathbf v \\
\mathbf v^\intercal A \mathbf w &= \lambda_{\mathbf w} \mathbf v^\intercal \mathbf w \\

(\mathbf w^\intercal A \mathbf x)^\intercal &= (\lambda_{\mathbf v} \mathbf w^\intercal \mathbf v)^\intercal \\
\mathbf v^\intercal A^\intercal \mathbf w &= \lambda_{\mathbf v} \mathbf w^\intercal \mathbf v = \lambda_{\mathbf v} \mathbf v^\intercal \mathbf w \qquad (\text{right side is just a scalar}) \\
0 &= (\lambda_{\mathbf w} - \lambda_{\mathbf v}) \mathbf v^\intercal \mathbf w \qquad \cdots \text{(5)(7)} \\
0 &= \mathbf v^\intercal \mathbf w \qquad \because \lambda_{\mathbf w} \neq \lambda_{\mathbf v} \\
&\therefore \mathbf v \perp \mathbf w
\end{align}
$$

### Recap (Symmetric matrix's orthogonal eigen vectors)

Let $A$ be a symmetric matrix.

$Q$ is orthogonal matrix. If we normalize the eigen vectors like $\dfrac{\mathbf v}{||\mathbf v||}$, then it is orthonormal.

$$
\begin{align*}
A
&= Q \Lambda Q^{-1} \\
&= Q \Lambda Q^\intercal \qquad \text{(orthogonal)} \\
&=
\begin{bmatrix}
\vert     & \vert     & \vert     \\
\mathbf v & \mathbf v & \mathbf v \\
\vert     & \vert     & \vert     \\
\end{bmatrix}
\Lambda
\begin{bmatrix}
\text{---} & \mathbf v^\intercal & \text{---} \\
\text{---} & \mathbf v^\intercal & \text{---} \\
\text{---} & \mathbf v^\intercal & \text{---} \\
\end{bmatrix}
\end{align*}
$$

1. $Q^\intercal$ part aligns the orthogonal eigen vectors to the original basis.
1. $\Lambda$ part scales each basis.
1. $Q$ part recovers the to the original basis.

### Algebraic representation of SVD

SVD as sum of rank 1 matrices (outer product).

$$
\begin{align*}
A &= U \Sigma V^\intercal \\
&= \sum_{i=1}^{r} \sigma_i \mathbf u_i \mathbf v_i^\intercal
= \sum_{i=1}^{r} \sigma_i \mathbf u_i \otimes \mathbf v_i
= \sum_{i=1}^{r} 
\sigma_i
\begin{bmatrix}
& \mathbf v_i & \\
\mathbf u_i \\
\\
\end{bmatrix}
\end{align*}
$$

```python
# low rank approximation

import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
import requests

A = Image.open(requests.get('https://picsum.photos/200/300', stream=True).raw).convert('L')
A = np.array(A, dtype=float)

fig, axs = plt.subplots(1,4, figsize=(12, 4), layout='constrained')
axs[0].imshow(A, cmap='binary_r')
axs[0].set_title(f"{A.shape = }")

U, S, Vh = np.linalg.svd(A)
S = np.pad(np.diag(S), ((0, 100), (0, 0)))

axs[1].imshow(U, cmap='binary_r')
axs[1].set_title(f"{U.shape = }")

axs[2].imshow(np.log1p(S), cmap='binary_r')
axs[2].set_title(f"{S.shape = }")

axs[3].imshow(Vh, cmap='binary_r')
axs[3].set_title(f"{Vh.shape = }")
plt.show()

rank = 6
fig, axs = plt.subplots(1, rank+1, figsize=((rank+1)*2, 3), layout='constrained')
A_ = np.zeros_like(A)
for i in range(rank):
    temp = S[i,i] * np.outer(U[:,i], Vh[i,:])
    A_ += temp
    axs[i].imshow(temp, cmap='binary_r')
    axs[i].set_title(r"$\sigma \mathbf{u} \mathbf{v}^\intercal$")
    axs[i].set_axis_off()

axs[i+1].imshow(A_, cmap='binary_r')
axs[i+1].set_title(r"$\sum \sigma \mathbf{u} \mathbf{v}^\intercal$")
axs[i+1].set_axis_off()
plt.show()
```

### Variance

Let $X$ be an arbitrary RV (on an axis).

The most familiar property of a distribution is its **mean**, or **exptected value**, denoted by $\mu$. \
Defined as
$$
\mathbb E[X] \triangleq \sum_x x \ p_X(x) \qquad \mathbb E[X] \triangleq \int_x x \ f_X(x) dx
$$

Think of $p_X(x)$ as $\frac{k}{n}$. \
$x$ can have same values $k$-multiple times, they are all counted independantly to calculate the expectation.

The spread can be calculated as $(x-\mu)^2$. The expected spread is called a **variance**, denoted by $\sigma^2$. \
Defined as
$$
\mathrm{var}[X] \triangleq \mathbb E [(X-\mu)^2] = \mathbb E [X^2] - \mu^2
$$

...and they are all scalars.

We can also derive the useful result $\mathbb E [X^2] = \mu^2 + \sigma^2$

### Covariance (and Correlation)

$X$: random variable\
$\mathbf X$: random vector\
$\mathrm K_{\mathbf X \mathbf X}$: auto-covariance matrix\
$\mathrm K_{\mathbf X \mathbf Y}$: cross-covariance matrix

$$
\mathbf X = (X_1, X_2, \cdots, X_n) \quad \cdots \text{column vector} \\[20pt]

\begin{align*}
\mathrm K_{X_iX_j} &= \mathrm{cov}[X_i, X_j] \\
&= \mathbb E [(X_i - \mu_i)(X_j - \mu_j)] \\[20pt]

\mathrm K_{\mathbf X \mathbf X} &= \mathrm{cov}[\mathbf X, \mathbf X] = \mathrm{var}[\mathbf X] \\
&= \mathbb E [(\mathbf X - \mathbb E [\mathbf X])(\mathbf X - \mathbb E [\mathbf X])^{\mathsf T}] \\[20px]
&=
\begin{bmatrix}
\mathbb E [(X_1 - \mu_1)(X_1 - \mu_1)] & \mathbb E [(X_1 - \mu_1)(X_1 - \mu_2)] & \cdots & \mathbb E [(X_1 - \mu_1)(X_1 - \mu_n)] \\[5px]
\mathbb E [(X_2 - \mu_2)(X_1 - \mu_1)] & \mathbb E [(X_2 - \mu_2)(X_1 - \mu_2)] & \cdots & \mathbb E [(X_2 - \mu_2)(X_1 - \mu_n)] \\[5px]
\vdots & \vdots & \ddots & \vdots \\[5px]
\mathbb E [(X_n - \mu_n)(X_1 - \mu_1)] & \mathbb E [(X_n - \mu_n)(X_1 - \mu_2)] & \cdots & \mathbb E [(X_n - \mu_n)(X_1 - \mu_n)] \\[5px]
\end{bmatrix} \\[80pt]

\mathrm R_{X_iX_j} &= \mathrm{corr}[X_i, X_j]
= \frac{\mathrm{cov}[X_i, X_j]}{\sqrt{\mathrm{var}[X_i] \ \mathrm{var}[X_j]}}
= \frac{\mathrm{cov}[X_i, X_j]}{\mathrm{\sigma}[X_i] \ \mathrm{\sigma}[X_j]} \\[20pt]

\mathrm R_{\mathbf X \mathbf X} &= \mathrm{corr}[\mathbf X, \mathbf X]
=
\mathrm{diag}(\mathrm{K}_{\mathbf X \mathbf X})^{-\frac{1}{2}} \
\mathrm{K}_{\mathbf X \mathbf X} \
\mathrm{diag}(\mathrm{K}_{\mathbf X \mathbf X})^{-\frac{1}{2}} \\[20pt]
&=
\begin{bmatrix}
\dfrac{\mathbb E [(X_1 - \mu_1)(X_1 - \mu_1)]}{\sigma(X_1)\sigma(X_1)} & \dfrac{\mathbb E [(X_1 - \mu_1)(X_1 - \mu_2)]}{\sigma(X_1)\sigma(X_2)} & \cdots & \dfrac{\mathbb E [(X_1 - \mu_1)(X_1 - \mu_n)]}{\sigma(X_1)\sigma(X_n)} \\[15px]
\dfrac{\mathbb E [(X_2 - \mu_2)(X_1 - \mu_1)]}{\sigma(X_2)\sigma(X_1)} & \dfrac{\mathbb E [(X_2 - \mu_2)(X_1 - \mu_2)]}{\sigma(X_2)\sigma(X_2)} & \cdots & \dfrac{\mathbb E [(X_2 - \mu_2)(X_1 - \mu_n)]}{\sigma(X_2)\sigma(X_n)} \\[15px]
\vdots & \vdots & \ddots & \vdots \\[5px]
\dfrac{\mathbb E [(X_n - \mu_n)(X_1 - \mu_1)]}{\sigma(X_n)\sigma(X_1)} & \dfrac{\mathbb E [(X_n - \mu_n)(X_1 - \mu_2)]}{\sigma(X_n)\sigma(X_2)} & \cdots & \dfrac{\mathbb E [(X_n - \mu_n)(X_1 - \mu_n)]}{\sigma(X_n)\sigma(X_n)} \\[15px]
\end{bmatrix} \\[50pt]

\end{align*}
$$

When a matrix is diagonal it is relatively easy to perform some calculations.

$$
\begin{bmatrix}
a & & \\
& b & \\
& & c \\
\end{bmatrix}^{x}
=
\begin{bmatrix}
a^x & & \\
& b^x & \\
& & c^x \\
\end{bmatrix}
$$

### MVN (Multivariative Normal)

$$
\begin{align*}
p(x) &= \mathcal N(x; \mu, \sigma^2)\\
&\triangleq \frac{1}{\sqrt{2\pi} \sigma} \exp \left[-\frac{1}{2}\left(\frac{x - \mu}{\sigma}\right)^2 \right]\\
\end{align*}
$$


$$
\begin{align*}
p(\mathbf x) &= \mathcal N(\mathbf x; \bm \mu, \mathbf \Sigma)\\
&\triangleq \frac{1}{(2\pi)^{D/2}|\mathbf \Sigma|^{1/2}} \exp \left[-\frac{1}{2}(\mathbf x - \bm \mu)^\intercal \mathbf \Sigma^{-1} (\mathbf x - \bm \mu) \right]\\
\end{align*}
$$

### Mahalanobis distance

* Euclidean distance
$$\{(\mathbf x - \bm \mu)^\intercal (\mathbf x - \bm \mu)\}^{1/D}$$

* Mahalanobis distance (for each cluster $i$)
$$\{(\mathbf x - \bm \mu)^\intercal \mathbf \Sigma_i^{-1} (\mathbf x - \bm \mu)\}^{1/D}$$

```python
import numpy as np
import matplotlib.pyplot as plt

C1 = np.array([[1, 3, 3, 5],
               [2, 1, 3, 2]])
C2 = np.array([[6, 8, 8,10],
               [6, 5, 7, 6]])

mu1 = C1.mean(1, keepdims=True)
mu2 = C2.mean(1, keepdims=True)
x  = np.array([[8],
               [2]])

plt.figure(figsize=(6,6))
plt.scatter(*C1, c='b', edgecolors='b', s=100)
plt.scatter(*C2, c='w', edgecolors='b', s=100)
plt.scatter(*mu1, c='b', edgecolors='b', marker='X')
plt.scatter(*mu2, c='w', edgecolors='b', marker='X')
plt.annotate(r'$\mathbf{\mu}_1$', xy=mu1, xytext=mu1+.2)
plt.annotate(r'$\mathbf{\mu}_2$', xy=mu2, xytext=mu2+.2)

plt.scatter(*x, c='orange', s=100)
plt.annotate(r'$\mathbf{x}_{\mathrm{unknown}}$', xy=x, xytext=x+.2)

plt.axis('square')
plt.xticks(range(0,13))
plt.yticks(range(0,13))
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.grid(alpha=.4)
plt.show()
```

```python
D = x.shape[0]

print('Euclidean')
print(((x - mu1).T @ (x - mu1))**(1/D), ((x - mu2).T @ (x - mu2))**(1/D))
print()

S1 = np.cov(C1)
S1_inv = np.linalg.inv(S1)

S2 = np.cov(C2)
S2_inv = np.linalg.inv(S2)

print('Mahalanobis')
print(((x - mu1).T @ S1_inv @ (x - mu1))**(1/D), ((x - mu2).T @ S2_inv @ (x - mu2))**(1/D))
```

Let $\mathbf x$ be a MVN RV of $D=2$.

$
\bm \mu =
\begin{bmatrix}
0\\
1
\end{bmatrix}
$

$
\mathbf \Sigma =
\begin{bmatrix}
1 & -0.75 \\
-0.75 & 2
\end{bmatrix}
$

```python
import numpy as np
from scipy.stats import multivariate_normal
import matplotlib.pyplot as plt

mean = [0, 1]
cov = [[ 1.00, -0.75],
       [-0.75,  2.00]]

l = np.linspace(-4, 4, 801)
X, Y = np.meshgrid(l, l)

rv = multivariate_normal(mean, cov)

Z = rv.pdf(np.stack([X, Y], axis=2))

plt.title(r"$\mathcal{N}(\mathbf{x};\mathbf{\mu}, \mathbf{\Sigma})$")
plt.contourf(X,Y,Z)
plt.scatter(*mean, c='r', marker='x')
plt.xlabel(r"$x_1$")
plt.ylabel(r"$x_2$")
plt.axis('square')
plt.show()
```

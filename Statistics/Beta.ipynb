```python
from scipy.stats import beta, binom
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 1, 101)

beta_pdf = lambda a, b, x: beta.pdf(a=a, b=b, x=x)

plt.plot(X, list(map(lambda x: beta_pdf(2, 5, x), X)), label='beta')
plt.xlim(0, 1)
plt.ylim(0)
plt.legend()
plt.show()

n = 40
plt.plot(range(n+1), [binom.pmf(n=n, k=i, p=.2) for i in range(n+1)])
plt.xlim(0, n)
plt.ylim(0)
plt.legend()
plt.show()
```

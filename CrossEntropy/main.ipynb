```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import matplotlib.pyplot as plt
import numpy as np
```

```python
count = 21
L = torch.linspace(-2, 2, count)
grid_X, grid_Y = torch.meshgrid(L, L)
softmaxed_X, softmaxed_Y = F.softmax(torch.concat([
    grid_X.ravel().unsqueeze(1),
    grid_Y.ravel().unsqueeze(1),
], dim=1), dim=1).T
softmaxed_X = softmaxed_X.reshape(count, count)
softmaxed_Y = softmaxed_Y.reshape(count, count)
```

```python
plt.figure(figsize=(8, 8))
plt.scatter(grid_X.flatten(), grid_Y.flatten(), marker='.')
plt.axis('square')
plt.axvline(0, lw=1, c='k')
plt.axhline(0, lw=1, c='k')
plt.scatter(softmaxed_X.flatten(), softmaxed_Y.flatten(), marker='.')
plt.xlim(-0.25, 1.25)
plt.ylim(-0.25, 1.25)
```

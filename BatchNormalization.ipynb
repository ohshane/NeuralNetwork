```python
import torch
import torch.nn as nn
```

```python
bn = nn.BatchNorm1d(1, affine=True)
X = torch.linspace(0, 4, 101)
sin = (lambda x: (x + (x*2*torch.pi).sin()))

import matplotlib.pyplot as plt

plt.figure(figsize=(8,8))
plt.plot(X, sin(X), label="sin")
plt.plot(X, bn(sin(X).unsqueeze(1)).detach(), label="bn")
plt.axhline(c='k', lw=1)
plt.legend()
plt.show()

bn._parameters
```

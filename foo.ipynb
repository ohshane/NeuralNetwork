```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(-2, 4, 13).reshape(-1,1)
X_Ones = np.concatenate([np.ones_like(X), X], axis=1)

Beta = np.array([[1.0],
                 [0.5]])

Epsilon = np.random.randn(13, 1)*.4

Y = X_Ones @ Beta + Epsilon
plt.scatter(X.T, Y.T)
plt.axvline(0, c="k", lw=1, alpha=.2)
plt.axhline(0, c="k", lw=1, alpha=.2)
plt.plot(X, X_Ones @ Beta, c="k", lw=1, ls="--")
plt.xlim(-3, 5)
plt.ylim(-3, 5)
plt.axis("equal")
plt.show()
```

```python
w = np.random.randn(2).reshape(-1, 1)
w_hat = np.random.randn(2).reshape(-1, 1)
w_tilde = (w - w_hat)
W = w_tilde @ w_tilde.T
print(w_tilde)
print(W)
```

```python
import torch
import torch.nn as nn
import torch.functional as F

import matplotlib.pyplot as plt
from mpl_toolkits import mplot3d


import numpy as np
```

```python
x = torch.randn(2, 100)
plt.scatter(*x, cmap='viridis_r', c=x[0]**2+x[1]**2)
plt.show()
```

```python
fig = plt.figure(figsize=(12, 12))
ax = plt.axes(projection='3d')


x = torch.randn(2, 100)
ax.scatter(x[0], x[1], cmap='viridis', c=sum(map(lambda x: x**2, x)), marker='o')

A = torch.randn(3, 2)
x_ = (A @ x)

ax.quiver3D(0,0,0,x_[0],x_[1],x_[2], alpha=.5, color='grey', linewidths=.5)
ax.scatter3D(x_[0],x_[1],x_[2], c='grey')

ax.set_xlim([-3, 3])
ax.set_ylim([-3, 3])
ax.set_zlim([ 0, 5])

plt.show()
```

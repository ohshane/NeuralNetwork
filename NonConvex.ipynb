```python
import torch
import torch.nn as nn
import torch.functional as F
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
l = torch.linspace(-2, 2, 41)
X, Y = torch.meshgrid(l, l, indexing='ij')

def rosenbrock(x, y):
    # https://en.wikipedia.org/wiki/Test_functions_for_optimization
    return (1 - x) ** 2 + 100 * (y - x ** 2) ** 2

Z = rosenbrock(X, Y)

fig = plt.figure(figsize=(16,8))
fig.suptitle('Rosenbrock')

ax = fig.add_subplot(121)
ax.contourf(X, Y, Z, cmap='terrain', alpha=.7, levels=[i**4 for i in range(10)])
ax.axvline(0, linewidth=1, c='k')
ax.axhline(0, linewidth=1, c='k')
ax.axvline(1, linewidth=1, linestyle='--', c='k')
ax.axhline(1, linewidth=1, linestyle='--', c='k')
ax.annotate('global minima (1,1)', [1,1])
ax.axis('square')

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, Z, cmap='terrain', alpha=.7)
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.view_init(30, 45)
plt.show()
```

```python
l = torch.linspace(-8, 8, 161)
X, Y = torch.meshgrid(l, l, indexing='ij')

def rastrigin(x, y):
    # https://en.wikipedia.org/wiki/Test_functions_for_optimization
    A = 10
    f = (
        A * 2
        + (x ** 2 - A * (x * torch.pi * 2).cos())
        + (y ** 2 - A * (y * torch.pi * 2).cos())
    )
    return f

Z = rastrigin(X, Y)

fig = plt.figure(figsize=(16,8))
fig.suptitle('Rastrigin')

ax = fig.add_subplot(121)
ax.contourf(X, Y, Z, cmap='terrain', alpha=.7)
ax.axvline(0, linewidth=1, c='k')
ax.axhline(0, linewidth=1, c='k')
ax.annotate('global minima (0,0)', [0,0])
ax.axis('square')

ax = fig.add_subplot(122, projection='3d')
ax.plot_surface(X, Y, Z, cmap='terrain', alpha=.7)
ax.view_init(30, -45)
plt.show()
```

```python
import torch
import torch.nn as nn
import torch.functional as F

import numpy as np
import matplotlib.pyplot as plt
```

```python
def gridplot(x_linspace, y_linspace):
    x = x_linspace
    y = y_linspace

    X, Y = np.meshgrid(x, y)

    for row in zip(X, Y):
        plt.plot(*row, 'g-', linewidth=1, alpha=.5)

    for row in zip(X.T, Y.T):
        plt.plot(*row, 'g-', linewidth=1, alpha=.5)

    plt.axis('equal')
    plt.axvline(0, c='black', linewidth=1)
    plt.axhline(0, c='black', linewidth=1)
```

```python
x = torch.linspace(-10, 10, 101)
y = torch.linspace(-10, 10, 101)

plt.figure(figsize=(12,6))
plt.suptitle('linear')

plt.subplot(121)
plt.plot(torch.linspace(-10, 10, 101), x)
plt.axvline(0, c='black', linewidth=1)
plt.axhline(0, c='black', linewidth=1)
plt.axis('equal')
plt.xlim([-4,4])
plt.ylim([-4,4])

plt.subplot(122)
gridplot(x,y)
plt.xlim([-4,4])
plt.ylim([-4,4])
plt.show()
```

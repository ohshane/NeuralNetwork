```python
x = np.linspace(1e-3, 1, 100)

plt.figure(figsize=(12,6))
plt.subplot(121)
plt.title(r'$-\log{p}$')
plt.plot(x, -np.log2(x))
plt.axhline(0, linewidth=1, c='k')
plt.axvline(0, linewidth=1, c='k')

plt.subplot(122)
plt.title(r'$-p\log{p}$')
plt.plot(x, -x*np.log2(x))
plt.axhline(0, linewidth=1, c='k')
plt.axvline(0, linewidth=1, c='k')
plt.show()
```

```python
# coin flip entropy

x = np.linspace(1e-3, 1-1e-3, 100)

plt.figure(figsize=(6,6))
plt.title('coin flip')
plt.plot(x, -x*np.log2(x) + -(1-x)*np.log2(1-x))
plt.ylabel('entropy')
plt.xlabel('proba.')
plt.axhline(0, linewidth=1, c='k')
plt.axvline(0, linewidth=1, c='k')
plt.show()
```

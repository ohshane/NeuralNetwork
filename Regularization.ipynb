```python
import torch
import torch.nn as nn
import torch.functional as F
import matplotlib.pyplot as plt
```

```python
l = torch.linspace(-10, 30, 401)
X, Y = torch.meshgrid(l, l)
X_ = X*4 + Y*-1
Y_ = Y*2 + X*-1

a, b = 20, 30

plt.figure(figsize=(24,8))
plt.suptitle(r'$L_1$')

plt.subplot(131)
plt.contourf(X, Y, X.abs()+Y.abs(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(132)
plt.contourf(X, Y, ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(133)
plt.contourf(X, Y, X.abs()+Y.abs() + ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')
plt.show()
```

```python
l = torch.linspace(-10, 30, 401)
X, Y = torch.meshgrid(l, l)
X_ = X*4 + Y*-1
Y_ = Y*2 + X*-1

a, b = 20, 30

plt.figure(figsize=(24,8))
plt.suptitle(r'$L_2$')

plt.subplot(131)
plt.contourf(X, Y, (X**2+Y**2).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(132)
plt.contourf(X, Y, ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(133)
plt.contourf(X, Y, (X**2+Y**2).sqrt() + ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')
plt.show()
```

```python
l = torch.linspace(-10, 30, 401)
X, Y = torch.meshgrid(l, l)
X_ = X*4 + Y*-1
Y_ = Y*2 + X*-1

a, b = 20, 30
p = 20

plt.figure(figsize=(24,8))
plt.suptitle(r'$L_p$')

plt.subplot(131)
plt.contourf(X, Y, (X.abs()**p+Y.abs()**p).pow(1/p), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(132)
plt.contourf(X, Y, ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')

plt.subplot(133)
plt.contourf(X, Y, (X.abs()**p+Y.abs()**p).pow(1/p) + ((X_-a)**2/16 + (Y_-b)**2/1).sqrt(), alpha=.3, levels=range(130))
plt.axis('square')
plt.axvline(0, linewidth=1, c='k')
plt.axhline(0, linewidth=1, c='k')
plt.show()
```

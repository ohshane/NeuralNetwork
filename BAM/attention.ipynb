```python
import torch
import torch.nn as nn
import torch.functional as F
import matplotlib.pyplot as plt
```

```python
x = torch.randn(32, 5, 5)

for i in range(len(x)):
    x[i].normal_((1.02)**i, 1)

avgpool = nn.AdaptiveAvgPool2d(1)
avgpool(x).flatten()

maxpool = nn.AdaptiveMaxPool2d(1)
maxpool(x).flatten()
```

```python
plt.plot(avgpool(x).flatten(), label="avgpool")
plt.plot(maxpool(x).flatten(), label="maxpool")
plt.legend()
plt.show()
```
